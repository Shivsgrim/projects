# **Importing the required libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

# **Reading the dataset**

```python
gas_data = pd.read_csv("encoded_feature.csv", index_col="Date", parse_dates=True)
```

```python
print(gas_data)
print("\nDescription of data: ")
print(gas_data.describe())
print("\nNumber of rows and columns:", gas_data.shape)
```

## Printing the first five rows of the dataset by using the head command

```python
gas_data.head()
```

# **Data Trends**

## Line plot shows the trend or the frequency of the data(price) along another factor(year). We can infer how the data changes over the time period

```python
plt.figure(figsize=(16, 6))
plt.title("Price of Natural gas in 1996 to 2010")
sns.lineplot(data = gas_data.Price)
plt.xlabel("Year")
plt.ylabel("Price in USD")
```

We can see that price of natural gas has been reduced on an average unlike the years from 2000 to 2010

checking for any null values in the data

```python
gas_data.isnull().sum()
```

so the price column has a single null value

# **Data relationship**

## Bar plot represents an estimate of central tendency for a numeric variable with the height of each rectangle

```python
plt.figure(figsize = (14, 7))
plt.title("Price of Natural gas month wise")
sns.barplot(x=gas_data["Months"], y=gas_data["Price"])
```

From this bar plot we can infer that the price of natural gas over the years has a tendency to be lower in the months of Feb-Apr, Aug-Oct and to be higher in the months of Nov-Jan, May-July

## Swam plot is another way of plotting the distribution of an attribute or the joint distribution of a couple of attributes and is used for representing categorical values

```python
plt.figure(figsize = (14, 7))
sns.swarmplot(x=gas_data['Seasons'],
              y=gas_data['Price'])
```

We can see that price has hit the hightest values in winter and autumn seasons which implies that the use of natural gas is higher in winter and autumn.

# **Data Distribution**

## A Distplot or distribution plot, depicts the variation in the data distribution

```python
plt.figure(figsize = (14, 7))
sns.distplot(gas_data["Price"])
```

We can see that the price of natural gas from 2 USD to 5 USD has higher density. That is for many years the price of natural gas has been in that range

## Jointplot is used to quickly visualize and analyze the relationship between two variables and describe their individual distributions on the same plot.

```python
plt.figure(figsize = (14, 7))
sns.jointplot(x=gas_data['Price'], y=gas_data.index, kind="kde")
```

It is similar to the displot, but here we can also see the years for which the prices where repeated over the years

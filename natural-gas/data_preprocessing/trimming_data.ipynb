```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("nasdaq_.csv")
```

```python
df["Date"] = pd.to_datetime(df["Date"])
df.info()
```

```python
df = df.set_index("Date")
df.head()
```

```python
df.info()
```

```python
df[df.index.month == 1]
```

```python
new_df = pd.DataFrame()

for i in range(1997, 2020+1):
    year_data = df[df.index.year == i]
    for j in range(1, 13):
        monthly_data = year_data[year_data.index.month == j]
        if not monthly_data.empty:
            new_df = new_df.append(monthly_data[monthly_data.Price == max(monthly_data.Price)])

new_df = new_df.drop_duplicates()
```
